## Day 3 - Data Cleaning 

## Objectives
- Standardize Ship-State naming conventions.
- Handle missing values in Qty and Amount.
- Verify row integrity after cleaning.

```python
## Import Libraries

import pandas as pd
import numpy as np
```

```python
## Dataset 
df = pd.read_csv(r"C:\Users\SVS\Desktop\data-analysis-project\data\raw\Amazon Sale Report.csv")
```

```python
df.shape
```

## Standardizing Ship-State 

```python
df['ship-state'].value_counts().sort_index()
```

```python
df['ship-state'] = (df['ship-state'] .astype(str).str.strip().str.title())
```

```python
df['ship-state'].value_counts().sort_index()
```

```python
## replacing correct values 
df['ship-state'] = df['ship-state'].replace({
    'Orissa': 'Odisha',
    'Rajshthan': 'Rajasthan',
    'Rajsthan': 'Rajasthan',
    'Rj': 'Rajasthan',
    'Pb': 'Punjab',
    'Pondicherry': 'Puducherry',
    'New Delhi': 'Delhi',
    'Punjab/Mohali/Zirakpur': 'Punjab',
    'Nl': 'Nagaland',
    'Ar': 'Arunachal Pradesh',
    'Nan': np.nan,
    'Apo': np.nan
})
```

```python
## Remove Spaces 
df['ship-state'] = df['ship-state'].str.strip()
```

```python
df['ship-state'].value_counts()
```

## Impute missing values for 'Qty' and 'Amount' columns


```python
## check missing values 

df[['Qty', 'Amount']].isnull().sum()
```

```python
df[['Qty', 'Amount']].describe()
```

## Filling Missing Values 

```python
## for quqntity 
df['Qty'] = df['Qty'].fillna(df['Qty'].median())
```

```python
df[['Qty', 'Amount']].isnull().sum()
```

```python
## for Amount 
df['Amount'] = df['Amount'].fillna(df['Amount'].median())
```

```python
df[['Qty', 'Amount']].isnull().sum()
```

##  Observation

The missing values in the Qty  and  Amount  columns were replaced using the  median . The median was chosen because the Amount column has a positively skewed distribution with several high-value outliers. Using the median preserves the dataset while minimizing the impact of extreme values.
